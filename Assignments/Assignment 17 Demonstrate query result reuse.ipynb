<div class="question--question--152Yi"><div class="ud-text-bold question--title--2FD-U"><div data-purpose="safely-set-inner-html:rich-text-viewer:html" class="question--title-text--2kBEO rt-scaffolding"><p>How can Snowflake return query results without requiring a virtual warehouse to be active?</p></div></div><div class="question--answer--2ASXa"><div data-purpose="safely-set-inner-html:submission:user-answer-body"><p>Query Result Cache within the Cloud Services layer stores results for executed queries. If a new query matches a prior one, Snowflake uses the cache to return the results without needing to re-execute a query. No active virtual warehouse is required, as the results are returned from the cloud services layer.</p></div></div></div>
