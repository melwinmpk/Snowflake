<div class="practice--panel--ejs8J"><div data-purpose="practice-component-text" class="rt-scaffolding"><h4>Lab Objective: Explore Snowflake Marketplace &amp; Consume Data</h4><p><br></p><p>This hands-on activity guides you through the process of exploring the Snowflake Marketplace. The objective is for the student to familiarize themself with the Snowflake Marketplace interface and understand the typical flow when consuming data from Snowflake Marketplace.</p><p>Although any user or role can explore the Snowflake Marketplace, you will need a user with the ACCOUNTADMIN privilege or the IMPORT SHARE privilege for consuming data. For simplicity, we suggest you utilize a user with the ACCOUNTADMIN privilege.</p><p><strong>1. </strong>Let us start by logging into the Snowsight Web Interface. If you use the classic Web UI and attempt to access Snowflake Marketplace, the link will redirect you to Snowsight.</p><p><br></p><p><strong>2.</strong> Click Marketplace in the left Navigation tab, as shown below.</p><figure><img src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-14-f14aeb9b602d310953b9c50de98bb2fb.png"></figure><p><br></p><p><strong>3.</strong> You can zero in on specific data sets that match your requirements using a variety of methods. For example, you can use the search functionality provided by Snowflake to search the Snowflake Marketplace. Provide keywords related to the type of data you are looking for; for example, you can search for flight data, customer sentiment data, retail, etc. Try the search capability by searching for some keywords you can think of.</p><figure><img src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-15-3c52f6cc3d3a3a15e0b5b659a2af590e.png"></figure><p><br></p><p><strong>4.</strong> Another way to quickly zero in on specific data sets is to use the filter capability. When you navigate to the Marketplace user interface, frequently used filters and categories are provided immediately below the search bar. Some frequently used filters are "Free Data" and "Ready to Query."</p><figure><img src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-15-a4df3a1a92b73c78ca5fbb0074621646.png"></figure><p><br></p><p><strong>5.</strong> You can, of course, browse the data available based on the category. The "Browse Data Products" drop-down is next to the search bar, which shows all the categories you can browse. The most used categories are shown right at the top of the drop-down. Select any category that interests you to view the data sets available under that category.</p><figure><img width="244" src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-15-b6847169a038555d6e3d5af8f2a82748.png" height="338"></figure><p><br></p><p><strong>6.</strong> You can also browse data sets according to different data providers. So if you have a favorite or a known data provider, you can browse them using the data provider drop-down. Select any data provider to see all the data sets provided by them.</p><figure><img width="226" src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-15-1957563b3be4183c3961f920c831dac0.png" height="373"></figure><p><br></p><p><strong>7.</strong> Let us now search for a data set and import it to our Snowflake account. We will add the weather data provided by AccuWeather in this example, but you can choose any other data set and category that you fancy.</p><p><br></p><p><strong>8.</strong> From the "Browse Data Products" drop-down, select the "Weather" category. If you don't see it at the top, locate it under "More Categories."</p><figure><img width="269" src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-15-e003e48503aff12d67e9d0b3775da845.png" height="393"></figure><p><br></p><p><strong>9.</strong> Now we will narrow down our results. Under the filter, select Availability and click "Ready to Query." Selecting "this filter will limit the results to only those data sets which are readily available, and once you import them, you can start running queries against them immediately. The other data set availability type is "By Request &amp; Personalized," meaning you will need to request them, the provider will review your request, get in touch with you, and they will likely customize the data to your need. For this lab, click "Ready to Query."</p><figure><img width="367" src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-15-b89bc106b54d083d219a48e8d71b005d.png" height="148"></figure><p><br></p><p><strong>10. </strong>Next, we will limit ourselves to just the free datasets. To do so, select the Price drop-down and select "Free."</p><figure><img width="593" src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-15-dc51b444ba3575eaeaae2ff592de5470.png" height="233"></figure><p><br></p><p><br></p><p><strong>11.</strong> Explore the various data sets that are listed. You can choose any data set of your liking. We will choose the "Sample of AccuWeather's Historical Weather Data" for this lab.</p><figure><img width="224" src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-15-b82c4b23db5e7f99c818e06569ef19f5.png" height="283"></figure><p><br></p><p><strong>12.</strong> Clicking a dataset takes you to the details screen for that data set, which provides a variety of information, such as the description of the data provider and the data set, some examples of the business needs that this data set can help with, and the regions where the data set is available. The screen also provides a "Get" button which you can use to gain access to the dataset.</p><figure><img src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-15-4262ad49860c90aaf8d66f2febc3b6a2.png"></figure><p><br></p><p><strong>13.</strong> Scrolling down on the data set screen, you can see usage examples, typically one or more SQL statements that you can use to understand how to query the data.</p><figure><img width="820" src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-16-9577561b3caa51ba94cd180e0721d3f9.png" height="549"></figure><p><br></p><p><br></p><p><strong>14.</strong> Let us now click the Get button and gain access to this data set. Two settings are required on this screen; the name of the database that will be created for this shared data set and the roles that can access the database. Provide a name for the database and select the roles that you want to be able to access this database. In the example shown, we named the database ACCUWEATHER and made it available to all roles in the database by granting it to the PUBLIC role. Note: You need to perform this step using the ACCOUNTADMIN role or a role with the IMPORT SHARE privileges; otherwise, the operation will fail.</p><figure><img width="448" src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-16-a15607bd30a00a362fa75995586cec5d.png" height="492"></figure><p><br></p><p><br></p><p><br></p><p><strong>15.</strong> The following error will show if you don't have the correct role before performing this operation. You can overcome this by changing to the ACCOUNTADMIN role or a role with IMPORT SHARE privileges.</p><figure><img src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-16-a12c626ed7a6ad5c7ea661d79bb32f96.png"></figure><p><br></p><p><br></p><p><strong>16.</strong> Assuming that the import operation has gone correctly, you will see the following message.</p><figure><img width="231" src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-16-374f903e82da604554ae265e200b0c36.png" height="237"></figure><p><br></p><p><br></p><p><strong>17.</strong> If you click "Open," the worksheet view will open with the new database pre-selected and the sample SQL pre-loaded in the worksheet view so that you can immediately start querying the data.</p><figure><img src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-16-7b79cf0e6072aff492f8d60f7b92640c.png"></figure><p><br></p><p><br></p><p><strong>18.</strong> Go to the left navigation pane, select Databases, and select the ACCUWEATHER database. (you may have a different database name based on the configuration you provided). The database detail page shows that the database was created from Snowflake Marketplace and also shows the data provider under the source details.</p><figure><img src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-16-b0adac8e9e7ccd49eb6d33b7732aa3ff.png"></figure><p><br></p><p><br></p><p><strong>19.</strong> Explore the schemas available under the ACCUWEATHER database. Next, click the "HISTORICAL" schema (you might have a different schema depending on your chosen dataset). Then, navigate to the "Views" tab to see all the views this shared database provides.</p><figure><img src="https://img-c.udemycdn.com/redactor/raw/assignment/2022-12-04_08-37-16-fc624be930bbef57dbc8f0adce297266.png"></figure><p><br></p><p><br></p><p><strong>20.</strong> Now that the database has been imported and is available in the system, you can query the data just like any other table in your system.
