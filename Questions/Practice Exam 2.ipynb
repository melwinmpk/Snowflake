<p>Which of the following correctly describes the behaviour when a Temporary table is attempted to be cloned?</p>

<li>Temporary tables can be cloned to permanent tables.</li>
</ul>

<h4>Explanation</h4>
<p>
Temporary tables can NOT be cloned to a permanent table.<br>
Doing so will typically show the following error “Temp table cannot be cloned to a permanent table; clone to a transient table instead.”<br>
However, a temporary table may be cloned to a transient table or another temporary table.</p>
