<h2>Use Defined Functions</h2>
<ul>
<li>User Defined Functions (UDFs) can be created to create custom functionality</li>
<li>Supported Languages</li>
<ul>
    <li>SQL</li>
    <li>Java</li>
    <li>JavaScript</li>
    <li>Python</li>
</ul>
<li>Scalar UDfs return one single value for each input</li>
<ul>
    <li>E.g MAX(2,3) returns 3</li>
</ul>
<li>Table UDFs or Tabular UDFs (UDTFs) can return zero, one or several rows for an input. Each result row can have multiple columns</li>
<ul>
    <li>Eg Flatten</li>
</ul>
</ul>

<h3>External Functions</h3>
<ul>
<li>Type of UDFs</li>
<li>Doesn't have its own code</li>
<li>Code is stored and executed outside of Snowflake</li>
<li>Snowflake only maintains metadata</li>
<li>External functions can be written in languages thaat standardUDFs don't support</li>
<li>They can also access functions and libraries that typical UDF can't</li>
<li>They can of course make calls to external API services</li>
</ul>

<h3>Secure UDF</h3>
<ul>
    <li>Typical UDF uses SQL execution optmizations which can enable access to underlying data</li>
    <li>This may allow access to data to users indirectly</li>
    <ul>
        <li>Security concern</li>
    </ul>
    <li>Secure UDFs solve this by not using the optmization</li>
    <ul>
        <li>Thus, ensuring users have no access to underying data, even indirectly</li>
    </ul>
    <li>Users can also see SQL logic or JavaScript code behind a typical UDF</li>
    <ul>
        <li>you may want to hide the code for security reasons</li>
    </ul>    
    <li>Secure UDF solves this issue by hiding the definition</li>
    <ul>
        <li>Only available to authorized users</li>
    </ul>    
    <li>Performance vs Security</li>
</ul>

<h3>Store Procedure</h3>
<ul>
    <li>Procedural code with if-else logic, looping and other features</li>
    <li>Construct dynamic SQL statements and execute</li>
    <li>Store procedures can return a single value or a tabular data</li>
    <li>Not mandatory to return a value</li>
    <li>UDFs & Store Procedures appear similar but serve a different purpose</li>
    <li>UDF takes an input, performance computation on the input, and return a value</li>
    <li>Store Procedures executes one or more SQL queries</li>
    <ul>
        <li>They are often used for performing administrative tasks</li>
    </ul>
    <li>Caller's Right</li>
    <ul>
        <li>Use privileges of the caller</li>
    </ul>
    <li>Owner's Right</li>
    <ul>
        <li>Default</li>
        <li>Uses Privileges of the creating role</li>
    </ul>
    <li>Stored Procs can be written in</li>
    <ul>
        <li>SQL</li>
        <li>JavaScript</li>
        <li>Snowflake Scripting</li>
        <li>& when using Snowpark, you can use Java, Python or Scala</li>
    </ul>    
</ul>

<h3>Snowpark</h3>
<ul>
    <li>Snowpark API</li>
    <ul>
        <li>Provides programming language support for building SQL that is executed on the Server</li>
        <li>Programmers can create code in a familiar language</li>
    </ul>
    <li>The code gets translated to SQL and is executed on Snowflake</li>
    <ul>
        <li>Takes advantage of the Snowflake scale & parallelism</li>
    </ul>
    <li>The Snowpark API functions are evaluated in Lazy manner ie nothing is done until a function is called that performa an action</li>
    <li>Programming language functions are translated to UDFs</li>
    <li>Executed using a Snadbox</li>
</ul>

<h3>Snowflake Scripting</h3>
<ul>
    <li>Extension to SQL that allows procedural logic similar to other programming languages</li>
    <li>Snowflake Scripting allows</li>
    <ul>
        <li>Variables</li>
        <li>If Else expressions</li>
        <li>Loops</li>
        <li>Cursors</li>
        <li>Results sets</li>
        <li>Exception Management</li>
    </ul>
    <li>Snowflake Scripting can be used to create store procedured</li>
    <ul>
        <li>It may be used to create procedural code outside of store procedures</li>
    </ul>
    <li>Procedural code is written in a Block within a Snowflake Scripting</li>
    <li>A Snowflake Scripting block has the following structure</li>
    <ul>
        <li>
        <pre>
            DECLARE
            .. variables declaration etc
            BEGIN
            .. Body of the block containing logic and SQL statement
            EXCEPTION
            .. Exception handeling statement
            END;
        </pre>
        </li>
    </ul>    
    <li>DECLARE & Exception are optional</li>
    <li><b>Variables</b> are declared in the DECLARE section</li>
    <li>They can be used in</li>
    <ul>
        <li>Expressions</li>
        <li>Snowflake Scripting statements</li>
        <li>SQL statements</li>
    </ul>    
    <li>A Typical declaration & use of variable may look like the following</li>
    <ul>
        <li>
        <pre>
            DECARE
                revenue numbrt(38,2);
            BEGIN
                let revenue number(38,2) := 0.0;
            ....
        </pre>
        </li>
    </ul> 
    <li>Snowflake Scripting supports two types of branching constructs</li>
    <ul>
       <li>IF-THEN-ELSEIF-ELSE</li>
        <li>CASE</li>
    </ul>    
    <li>Like typical procedural languages Snowflake Scripting supports several types of loops</li>
    <ul>
       <li>FOR</li>
        <li>WHILE</li>
        <li>REPEAT</li>
        <li>LOOP</li>
    </ul>    
    <li>A Snowflake Scripting block can return a value using the RETURN command</li>
</ul>
