<h3>Data Encription at Rest</h3>
<ul>
    <li><h3>Data Encryption at Rest</h3></li>
    <ul>
        <li>Snowflake stores all customer data using AES-256 bit encryption</li>
        <li>snowflake managed keys are rotated every 30 days, and data is re-encrypted with new keys eery year(yearly re-keyinh) </li>
        <li>With Tri Secret Secure (required Business Critial edition) you can further protect your data with combining of snowflake managed and a customer-managed key</li>
    </ul>
    <li><h3>Authentication</h3></li>
    <ul>
        <li>Snowflake supports key pair authentication using private and public key</li>
        <li>Snowflake also supports federated authentication (SSO) where users authenticate via an external SAML 2.0 identity provider</li>
        <li>Snowflake supports SCIM an open standard for automatic user and role provisioning based on identity provider data</li>
        <li>Snowflake also enforces strict password rules</li>
        <li>All snowflake editions supports MFA or multifactor authentication</li>
        <li>Administrators can disable or temporary disable MFA for a user</li>
        <li>Snowflake Web UI , SnowSQL and all other connectors support MFA</li>
        <li></li>
    </ul>    
    <li><h3>Authorization</h3></li>
    <ul>
        <li>Snowflake's access controle system is based on RBAC or role-based access controle model</li>
        <li>Snowflake also offers <b>DAC or discretionary access control</b> Privileges can be granted on Securable objects</li>
        <li>The Grant statement grants privileges while the REVOKE statement revokes them</li>
        <li>A role is an entity to which privileges on other objects cna be granted</li>
        <li>The role themselves can then be granted to other users or other roles. A role hierarchy is formed when a role is assigned to another role</li>
        <li>Snowflake inclused a small set of system roles</li>
        <li>ACCOUNTADMIN is the full-access account administrator role</li>
        <li>USERADMIN allows you to create USERS and ROLES</li>
        <li>SECURITYADMIN inherits USERADMIN privileges and has the ability to handel object grnats globally</li>
        <li>Most item in your snowflake system can be created and managed using SYSADMIN</li>
        <li>PUBLIC is the lest permissive role and it is assigned to everyone by default</li>
        <li>When a role is assigned to another role, the higher role obtains all of the privileges of the lower role</li>
        <li>Snowflake's roles are arranged in a hierarchical order</li>
        <li><b>Column-level</b> security feature in Snowflake allowa you to mask a column</li>
        <li><b>Masking policies</b> are attached to columns to enable masking</li>
        <li>Two Methods of Masking</li>
        <ul>
            <li>Dynamic Data masking</li>
            <ul>
                <li>data stored unmasked</li>
                <li>show masked or unmasked data at execution time</li>
            </ul>    
            <li>External tokenization</li>
            <ul>
                <li>data stored in a tokenized (encrypted manner)</li>
                <li>show masked or inmasked data at execution time</li>
            </ul>    
        </ul>    
        <li>Snowflake also offers <b>row-level</b> security</li>
        <ul>
            <li>Determine which rows to return to users when they execute queries</li>
            <li><b>Row-level access policies</b> define which rows should be returend using conditions and functions</li>
        </ul>    
    </ul> 
    <li>Network</li>
    <ul>
        <li>Snowflake <b>encrypts</b> all <b>data in transit using TLS1.2</b></li>
        <ul>
            <li>Applies to all communication and all methods of connectivity</li>
            <li>Web interface, SnowSQL  and all the drivers</li>
        </ul>    
        <li>Administrators can define <b>network policies</b> to alow or denay <b>IPv4 addresses</b></li>
        <li>Network Policies have<b> allowed IP address & blocked IP addresses</b></li>
        <li>Snowflake applies the block list if both lists are populated</li>
        <li>Sowflake supports <b>private connectivity</b></li>
        <ul>
            <li>Requires <b>Business Critical edition and above</b></li>
        </ul>    
    </ul>
</ul>
